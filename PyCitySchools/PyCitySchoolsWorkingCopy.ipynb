```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete['MathPass'] = np.where(school_data_complete['math_score']>=70,1,0)
school_data_complete['ReadingPass'] = np.where(school_data_complete['reading_score']>=70,1,0)
school_data_complete['BothPass'] = np.where((school_data_complete['math_score']>=70) & (school_data_complete['reading_score']>=70),1,0)
school_data_complete
```

```python
total_percent_passing =  (len(school_data_complete.loc[school_data_complete["reading_score"]>=70]) & len(school_data_complete.loc[school_data_complete["math_score"]>=70]))/total_students
conv_pct_total ='{:.2f}%'.format(total_percent_passing * 100)
conv_pct_total
```
